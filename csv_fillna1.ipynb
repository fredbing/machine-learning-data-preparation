There is an invalid value (negative) in the Wind Speed column. it would make better sense to change it to na_values

```python
df1 = pd.read_csv("./fillna_data/weather_jan.csv", header=1, parse_dates=["Date"], na_values={'Wind Speed': -1})
df1.set_index('Date', inplace=True)
df1
```

```python
df2 = df1.fillna(0)
df2
```

#### it can be seen from above that fillna() function only fills the empty fileds, but not the fields having values as n.a. Hence it needs to convert the fields with value 'n.a.' to 'na_values' (NaN)

```python
df3 = pd.read_csv("./fillna_data/weather_jan.csv", header=1, parse_dates=["Date"], na_values={'High Temperature':'n.a.','Low Temperature':'n.a.','Wind Speed':['n.a.', -1], 'Precipitation':'n.a.'})
df3.set_index('Date', inplace=True)
df3
```

#### Note the difference between none and NaN for precipitation. 'none' means no precipitation, 'NaN' means info not available

### the foward fill method "ffill" is used below, by which the NA fields copy the value in the cell above. backward fill method "bfill" can also be used

```python
df4 = df3.fillna(method="ffill")
df4
```

```python
df5 = df3.fillna(method="bfill")
df5
```

#### In this case, backforward fill leaves 2 NaN unfilled. forward fill could be a better option

 Other fill method such as interpolate() can be an option to choose, espeially considering there are bigger gaps between certain days, e.g., 1/2 and 1/5.

```python
df6 = df3.interpolate()
df6
```

```python
df7 = df3.interpolate(method='time')
df7
```

Note that the NaN's in 'Precipitation' column cannot be filled with the interpolate method. For machine learning purpose, these values need to be converted to numerical values to represent each of its catogory. interpolate method is not a proper way to fill the NA's for these types of data even after they are converted to numerical values.
